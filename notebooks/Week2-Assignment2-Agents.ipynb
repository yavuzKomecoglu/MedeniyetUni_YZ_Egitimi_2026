# Alıştırma: ReAct Pattern ile Agent Geliştirme

## Genel Bakış

Bu alıştırmada, LangChain'in `create_agent()` yapısını kullanarak otonom (autonomous) Agent'lar geliştireceksiniz.

Ayrıca Agent'ların:

* ReAct Pattern (Düşün → Eyleme Geç → Gözlemle) yaklaşımını kullanmasını,
* Bir problemi çözene kadar birden fazla adım çalıştırmasını,
* Birden fazla araç (Tool) arasından doğru olanı seçmesini,
* Gerekli durumlarda birden fazla aracı sırasıyla kullanmasını

uygulamalı olarak deneyimleyeceksiniz.


---

# Alıştırma: Research Agent 🔍

## Amaç

Web araması (Search) ve matematiksel işlemler (Calculation) gerektiren soruları çözebilen bir Agent geliştirmek.

---

## Yapılacaklar

### 1. İki Adet Tool Oluşturun

#### - Search Tool

Web aramasını simüle eden bir araç oluşturun.

Örnek olarak önceden tanımlanmış sonuçlar döndürebilir.

Örneğin:

```python
{
    "population of tokyo":
        "Tokyo has a population of approximately 14 million",

    "capital of france":
        "The capital of France is Paris",

    "capital of japan":
        "The capital of Japan is Tokyo"
}
```

---

#### - Calculator Tool

Matematiksel işlemler yapabilen bir araç oluşturun.

Örneğin:

```python
2 + 2
```

veya

```python
14000000 * 2
```

gibi işlemleri çalıştırabilmelidir.

---

### 3. Agent Oluşturun

`create_agent()` kullanarak Agent oluşturun.

Agent aşağıdaki araçlara sahip olmalıdır:

* Search Tool
* Calculator Tool

---

### 4. Çok Adımlı Sorularla Test Edin

Agent'ın birden fazla Tool kullanmasını gerektiren sorular deneyin.

---

### 5. Kullanılan Araçları Gösterin

Agent'ın hangi araçları kullandığını ekrana yazdırın.

Örnek:

```text
Tools used:
search_tool
calculator_tool
```

---

## Örnek Sorular

### Soru 1

```text
What is the population of Tokyo multiplied by 2?
```

Beklenen süreç:

```text

2. Nüfusu 2 ile çarp
3. Sonucu döndür
```

### Soru 2

```text
Search for the capital of France and tell me how many letters are in its name
```

Beklenen süreç:

```text
1. Fransa'nın başkentini bul
2. Paris kelimesinin harf sayısını hesapla
3. Sonucu döndür
```

---

## Başarı Kriterleri

Çözümünüz aşağıdaki kriterleri karşılamalıdır:

* `create_agent()` kullanılmalı
* Her Tool açıklayıcı bir description içermeli
* Agent hangi Tool'u kullanacağına kendisi karar vermeli
* Çok adımlı problemleri çözebilmeli
* Kullanılan Tool'lar görüntülenebilmeli
* Doğru sonuçlar üretebilmeli
