# Week 2 - PART 3 - Agents - SQL Agent Oluşturma (Build a SQL Agent)


## Genel Bakış

Bu bölümde, LangChain kullanarak bir SQL veritabanı ile doğal dil üzerinden iletişim kurabilen bir AI Agent geliştireceğiz.

Amaç, kullanıcının SQL bilmesine gerek kalmadan sorular sorabilmesi ve Agent'ın bu soruları SQL sorgularına dönüştürerek veritabanından cevap üretebilmesidir.

Örneğin kullanıcı:

> En çok satış yapılan ülke hangisi?

veya

> Son bir ayda en fazla sipariş veren müşteriler kimler?

gibi sorular sorabilir.

Agent ise bu soruları anlayarak gerekli SQL sorgularını oluşturur, veritabanını sorgular ve sonuçları kullanıcıya anlaşılır bir şekilde sunar.

---

## Agent Arka Planda Nasıl Çalışır?

Bir SQL Agent genellikle aşağıdaki adımları takip eder:

### 1. Veritabanını Tanır

İlk olarak veritabanında hangi tabloların bulunduğunu inceler.

Örneğin:

- Customers
- Orders
- Products
- Invoices

gibi tabloları keşfeder.

### 2. İlgili Tabloları Belirler

Kullanıcının sorduğu soruya göre hangi tabloların kullanılacağını seçer.

Örneğin:

> En çok sipariş veren müşteri kim?

sorusu için `Customer` ve `Order` tablolarının gerekli olduğunu anlayabilir.

### 3. Tablo Yapılarını İnceler

Seçilen tabloların kolonlarını, veri tiplerini ve ilişkilerini analiz eder.

Örneğin:

- CustomerId
- OrderDate
- TotalAmount

gibi alanları öğrenir.

### 4. SQL Sorgusunu Oluşturur

Topladığı bilgiler doğrultusunda uygun SQL sorgusunu üretir.

```sql
SELECT CustomerName,
       COUNT(*) AS OrderCount
FROM Orders
GROUP BY CustomerName
ORDER BY OrderCount DESC
LIMIT 1;
```

### 5. Sorguyu Kontrol Eder

Sorguyu çalıştırmadan önce yaygın SQL hatalarına karşı tekrar gözden geçirir.

Örneğin:

- Yanlış kolon adı kullanılmış mı?
- Hatalı JOIN yapılmış mı?
- Eksik GROUP BY var mı?
- Hatalı filtreleme yapılmış mı?

gibi kontroller gerçekleştirir.

### 6. Sorguyu Çalıştırır

Oluşturduğu SQL sorgusunu veritabanında çalıştırır ve sonuçları alır.

### 7. Hataları Düzeltir

Sorgu sırasında hata oluşursa hata mesajını analiz eder ve sorguyu yeniden düzenleyerek tekrar dener.

Örneğin aşağıdaki hata ile karşılaşabilir:

```text
no such column: customer_name
```

Bu durumda Agent:

- Hata mesajını inceler,
- Tablo şemasını tekrar kontrol eder,
- Doğru kolon adını bulmaya çalışır,
- Yeni bir SQL sorgusu üretir.

### 8. Sonucu Kullanıcıya Sunar

Elde ettiği verileri teknik olmayan bir dil ile yorumlayarak kullanıcıya aktarır.

Örneğin:

> En fazla sipariş veren müşteri Ahmet Yılmaz'dır. Toplam 148 sipariş oluşturmuştur.

---

## Neden SQL Agent Kullanıyoruz?

Klasik yöntemlerde kullanıcıların SQL bilmesi gerekir.

SQL Agent sayesinde:

- İş birimleri SQL öğrenmek zorunda kalmaz.
- Teknik olmayan kullanıcılar veriye erişebilir.
- Raporlama süreçleri hızlanır.
- Veritabanı ile doğal dil üzerinden etkileşim kurulabilir.

Bu sayede kullanıcılar:

> Geçen ay en çok satan ürün hangisi?

veya

> Son üç ayda en yüksek ciroyu hangi şehir üretti?

gibi soruları doğrudan sorabilir.

## Güvenlik Uyarısı

SQL Agent'lar model tarafından üretilen sorguları çalıştırabilir.

Bu nedenle güvenlik konusu son derece önemlidir.

Agent'ın kullandığı veritabanı hesabına yalnızca ihtiyaç duyduğu yetkiler verilmelidir.

Mümkün olduğunca:

- Sadece okuma (**READ ONLY**) yetkisi kullanılmalı,
- Gereksiz tablo erişimleri kapatılmalı,
- Üretim veritabanına doğrudan erişim yerine kısıtlı kullanıcılar tercih edilmelidir.

Bu önlemler riski tamamen ortadan kaldırmasa da olası hataların ve güvenlik problemlerinin etkisini önemli ölçüde azaltır.

## 3. Veritabanı Etkileşimi İçin Araçlar

Agent'ın veritabanı ile iletişim kurabilmesi için çeşitli araçlar kullanılır.

---

### sql_db_list_tables

Veritabanındaki mevcut tabloları listeler.

```python
@tool
def sql_db_list_tables() -> str:
```

Örnek çıktı:

```text
Album, Artist, Customer, Employee, Genre, Track
```

---

### sql_db_schema

Belirtilen tabloların şema bilgilerini ve örnek kayıtlarını getirir.

```python
@tool
def sql_db_schema(table_names: str) -> str:
```

Örnek kullanım:

```python
sql_db_schema("Track, Genre")
```

---

### sql_db_query

Oluşturulan SQL sorgusunu veritabanında çalıştırır.

```python
@tool
def sql_db_query(query: str) -> str:
```

Sorgu hatalıysa hata mesajı döndürür.

---

### sql_db_query_checker

SQL sorgusunu çalıştırmadan önce kontrol eder ve olası hataları düzeltmeye çalışır.

```python
@tool
def sql_db_query_checker(query: str) -> str:
```

Kontrol ettiği bazı yaygın hatalar:

- NULL ile `NOT IN` kullanımı
- `UNION` / `UNION ALL` hataları
- `BETWEEN` kullanım hataları
- Veri tipi uyuşmazlıkları
- Tanımlayıcı (identifier) yazım hataları
- Hatalı JOIN ilişkileri

> SQL sorgularının çalıştırılmadan önce bu araç ile kontrol edilmesi önerilir.

`````python
import sqlite3
from langchain.tools import tool

# Below are minimal tools for demonstration purposes.
# They are not intended to be secure or for production use.

@tool
def sql_db_list_tables() -> str:
    """Input is an empty string, output is a comma-separated list of tables in the database."""
    con = sqlite3.connect("Chinook.db")
    try:
        cursor = con.cursor()
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = [row[0] for row in cursor.fetchall() if not row[0].startswith("sqlite_")]
        return ", ".join(tables)
    finally:
        con.close()

@tool
def sql_db_schema(table_names: str) -> str:
    """Input to this tool is a comma-separated list of tables, output is the schema and sample rows for those tables.
    Be sure that the tables actually exist by calling sql_db_list_tables first!
    Example Input: table1, table2, table3"""
    con = sqlite3.connect("Chinook.db")
    try:
        cursor = con.cursor()
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        valid_tables = {row[0] for row in cursor.fetchall() if not row[0].startswith("sqlite_")}
        results = []
        for table in table_names.split(","):
            table = table.strip()
            if table not in valid_tables:
                results.append(f"Error: table_names {{{table!r}}} not found in database")
                continue
            cursor.execute("SELECT sql FROM sqlite_master WHERE type='table' AND name=?;", (table,))
            schema_row = cursor.fetchone()
            if schema_row:
                results.append(schema_row[0])
                try:
                    quoted_table = '"' + table.replace('"', '""') + '"'
                    cursor.execute(f"SELECT * FROM {quoted_table} LIMIT 3;")
                    rows = cursor.fetchall()
                    if rows:
                        col_names = [description[0] for description in cursor.description]
                        results.append(f"/*\n3 rows from {table} table:\n" + "\t".join(col_names) + "\n" + "\n".join("\t".join(str(x) for x in row) for row in rows) + "\n*/")
                except Exception as e:
                    results.append(f"Error fetching sample rows: {e}")
        return "\n\n".join(results)
    finally:
        con.close()

@tool
def sql_db_query(query: str) -> str:
    """Input to this tool is a detailed and correct SQL query, output is a result from the database.
    If the query is not correct, an error message will be returned.
    If an error is returned, rewrite the query, check the query, and try again.
    If you encounter an issue with Unknown column 'xxxx' in 'field list', use sql_db_schema to query the correct table fields."""
    con = sqlite3.connect("Chinook.db")
    try:
        cursor = con.cursor()
        cursor.execute(query)
        res = cursor.fetchall()
        return str(res)
    except Exception as e:
        return f"Error: {e}"
    finally:
        con.close()

@tool
def sql_db_query_checker(query: str) -> str:
    """Use this tool to double check if your query is correct before executing it.
    Always use this tool before executing a query with sql_db_query!"""
    trigger_prompt = """{query}
Double check the sqlite query above for common mistakes, including:
- Using NOT IN with NULL values
- Using UNION when UNION ALL should have been used
- Using BETWEEN for exclusive ranges
- Data type mismatch in predicates
- Properly quoting identifiers
- Using the correct number of arguments for functions
- Casting to the correct data type
- Using the proper columns for joins

If there are any of the above mistakes, rewrite the query. If there are no mistakes, just reproduce the original query.

Output the final SQL query only.

SQL Query: """.format(query=query)

    response = model.invoke(trigger_prompt)
    return response.text.strip()

tools = [sql_db_list_tables, sql_db_schema, sql_db_query, sql_db_query_checker]

for tool in tools:
    print(f"{tool.name}: {tool.description}\n")
`````

## 4. create_agent Kullanımı

LangChain'in `create_agent()` fonksiyonu kullanılarak bir **ReAct** Agent oluşturulur.

**Sistem prompt'u:**

`````python
system_prompt = """
You are an agent designed to interact with a SQL database.
Given an input question, create a syntactically correct {dialect} query to run,
then look at the results of the query and return the answer. Unless the user
specifies a specific number of examples they wish to obtain, always limit your
query to at most {top_k} results.

You can order the results by a relevant column to return the most interesting
examples in the database. Never query for all the columns from a specific table,
only ask for the relevant columns given the question.

You MUST double check your query before executing it. If you get an error while
executing a query, rewrite the query and try again.

DO NOT make any DML statements (INSERT, UPDATE, DELETE, DROP etc.) to the
database.

To start you should ALWAYS look at the tables in the database to see what you
can query. Do NOT skip this step.

Then you should query the schema of the most relevant tables.
""".format(
    dialect="sqlite",
    top_k=5,
)
`````

**Agent oluşturma:**

`````python
from langchain.agents import create_agent


agent = create_agent(
    model,
    tools,
    system_prompt=system_prompt,
)
`````

## 5. Agent'i Çalıştırma (Run the Agent)

**Örnek soru:**

```python
question = "Which genre on average has the longest tracks?"
````

`````python
question = "Ortalama olarak en uzun parçalara sahip müzik türü hangisidir?"

for step in agent.stream(
    {"messages": [{"role": "user", "content": question}]},
    stream_mode="values",
):
    step["messages"][-1].pretty_print()
`````

### Adım 3: SQL Sorgusunu Oluşturur

```sql
SELECT Genre.Name,
       AVG(Track.Milliseconds) AS AverageLength
FROM Track
INNER JOIN Genre
    ON Track.GenreId = Genre.GenreId
GROUP BY Genre.Name
ORDER BY AverageLength DESC
LIMIT 5;
```
